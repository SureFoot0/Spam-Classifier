```python
import numpy as np
import pandas as pd
import spacy
from nltk.stem.snowball import SnowballStemmer
from umap import UMAP
from sklearn.cluster import KMeans
from sentence_transformers import SentenceTransformer
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
df = pd.read_csv("Words.csv")
df = df.drop(columns=["Rank", "Part of speech", "Frequency", "Dispersion"])
```

```python
df = pd.DataFrame(columns=[df["Word"]])
df["LABEL"] = None
df.head()
```

```python
Emails = pd.read_csv("Text_Emails.csv")
Emails.head()
```

```python
def PrepNewData(nlp, text, prediction, df):
    doc = nlp(text)
    lemmaDoc = [token.lemma_ for token in doc]
    PreparedData = dict.fromkeys(df.columns[:-1], 0)
    print(PreparedData)
    for token in lemmaDoc:
        try:
            PreparedData[token] += 1
            print("Token Added")
        except:
            pass
    PreparedData = {key: [value] for key, value in PreparedData.items()}
    PreparedData["LABEL"] = prediction
    return PreparedData
```

```python
df = pd.concat([df, pd.DataFrame(PrepNewData(nlp, Emails["text"][0], Emails["label"][0], df))], ignore_index = True)
```

```python
df_target = pd.DataFrame(data=df["Prediction"], columns=["Prediction"])
df_target.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, np.ravel(df_target), test_size=0.3, random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```
